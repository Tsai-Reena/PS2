```python
import os
import json
import re
from langchain_community.llms import HuggingFaceEndpoint

os.environ['HUGGINGFACEHUB_API_TOKEN'] = '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' #改為自己的API Token
repo_id = "mistralai/Mistral-7B-Instruct-v0.2"
llm_try = HuggingFaceEndpoint(repo_id=repo_id)


# 載入 sample_data.json 文件
with open('./sample_data.json', 'r', encoding='utf-8') as file:
    sample_data = json.load(file)
```

# LLM分析留言

```python
#sample_data

data = {}
total = []
every = []
for index, item in enumerate(sample_data, start=1):

    every = []
    every.append(index)
    every.append(item.get('店名', ''))
    every.append(item.get('評論者', ''))
    every.append(item.get('評分', ''))
    #total.append(every)
    comment = item.get('評論內容', '')
    txt = llm_try.invoke("請找出下面的opinion feature, 例如:口味，態度，配料，並且列出品項，換算成分數，以[商品名稱,評分]格式輸出一次，滿分100，只用中文"+comment)
    matches = re.findall(r'\[.*?\]', txt)

    # 手動解析匹配到的字符串

    for match in matches:
        temp = every.copy()
        # 去掉兩邊的括號
        match = match[1:-1]
        # 使用逗號分隔元素
        elements = match.split(',')
        # 去除多餘的空格，並將數字轉換為整數
        elements = [e.strip() for e in elements]
        if(len(elements) > 1):
            if(elements[1].isnumeric()):
                elements[1] = int(elements[1])
                #print(elements)
                temp.append(elements)
                total.append(temp)
    
                
#print(total)
               
```

# 備份輸出資料

```python
text_file = open("Output.txt", "w",encoding='utf8')

text_file.write(str(total))

text_file.close()
```
